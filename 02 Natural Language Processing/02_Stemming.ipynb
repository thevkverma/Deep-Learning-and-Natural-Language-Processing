## Stemming

```python
## Classification Problem
## Comments of product is positive review or negative review
words = ["eating","eats","eaten","writing","writes","programs","history","finally","finalized"]
```

#### PorterStemmer

```python
from nltk.stem import PorterStemmer
```

```python
stemming=PorterStemmer()
```

```python
for word in words:
  print(word+"---->"+stemming.stem(word))
```

```python
stemming.stem("congratulations")
```

```python
stemming.stem("sitting")
```

#### Regexp Stemmer

```python
from nltk.stem import RegexpStemmer
```

```python
reg_strmmer = RegexpStemmer('int$|s$|e$|able$', min=4)
```

```python
reg_strmmer.stem('int$')
```

```python
reg_strmmer.stem('eating')
```

```python
reg_strmmer.stem('ingeating')
```

#### Snowball Stemmer

```python
from nltk.stem import SnowballStemmer
```

```python
snowballsstemmer = SnowballStemmer('english')
```

```python
for word in words:
       print(word+"---->"+snowballsstemmer.stem(word))
```

```python
stemming.stem('fairly'),stemming.stem('sportingly')
```

```python
snowballsstemmer.stem('fairly'),snowballsstemmer.stem('sportingly')
```
