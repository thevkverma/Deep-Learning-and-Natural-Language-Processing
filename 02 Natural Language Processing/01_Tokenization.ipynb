```python
corpus = """
AI has received increased attention from the information systems (IS) research community in recent years.
There is, however, a growing concern that research on AI could experience a lack of cumulative building of knowledge,
which has overshadowed IS research previously. This study addresses this concern,
by conducting a systematic literature review of AI research in IS between 2005 and 2020.
The search strategy resulted in 1877 studies, of which 98 were identified as primary studies and a synthesise of key themes that are pertinent to this study is presented.
In doing so, this study makes important contributions, namely (i) an identification of the current reported business value and contributions of AI,
(ii) research and practical implications on the use of AI and (iii) opportunities for future AI research in the form of a research agenda.
"""
```

```python
print(corpus)
```

```python
## Tokenization
## Sentence --> Paragraphs
from nltk.tokenize import sent_tokenize
```

```python
sent_tokenize(corpus)
```

```python
documents = sent_tokenize(corpus)
```

```python
type(documents)
```

```python
for sentence in documents:
  print(sentence)
```

```python
## Tokenization
## Paragraph --> Words
## Sentence --> words
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
for sentence in documents:
  print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus) ## in this case "'" and 's' both will be tokenize.
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer = TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```
