#### Sentiment Analysis

```python
from tensorflow.keras.datasets import imdb
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN, Embedding, Flatten
```

```python
(X_train, y_train),(X_test, y_test) = imdb.load_data()
```

```python
X_train[0]
```

```python
print(len(X_train[2]))
print(len(X_train[3]))
```

```python
X_train = pad_sequences(X_train, padding = 'post', maxlen = 50)
X_test = pad_sequences(X_test, padding = 'post', maxlen = 50)
```

```python
X_train[0]
```

```python
model = Sequential()

model.add(SimpleRNN(32, input_shape = (50,1), return_sequences=False))
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
model.fit(X_train, y_train, epochs = 5, validation_data=(X_test, y_test))
```

#### How to do encodings using keras embeddings

```python
from tensorflow import keras
from keras.datasets import imdb
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Embedding, SimpleRNN, Dense
from keras.optimizers import Adam
```

```python
# Load the IMDB dataset, limiting the vocabulary to 10000 words
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=10000)
```

```python
# Pad the sequences to a maximum length of 50
X_train = pad_sequences(X_train, padding='post', maxlen=50)
X_test = pad_sequences(X_test, padding='post', maxlen=50)
```

```python
# Define the model
model = Sequential()
# The Embedding layer is configured with a vocabulary size of 10000
model.add(Embedding(10000, output_dim=2, input_shape=(50,)))
model.add(SimpleRNN(32, return_sequences=False))
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
# Compile the model
model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['acc'])

# Train the model
history = model.fit(X_train, y_train, epochs=5, batch_size=32, validation_data=(X_test, y_test))
```

###### Predictions

```python
test_data = X_test[0][0:50].reshape(1,-1)
```
